# Sentiment Analysis

```python
from textblob import TextBlob
import matplotlib.pyplot as plt

# Function to get sentiment polarity
def get_sentiment(text):
    if pd.isna(text):
        return None
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis
ad['Description Sentiment'] = ad['Ticket Description'].apply(get_sentiment)
ad['Resolution Sentiment'] = ad['Resolution'].apply(get_sentiment)

# Plotting the sentiment distribution
plt.figure(figsize=(10, 5))
plt.hist([ad['Description Sentiment'].dropna(), ad['Resolution Sentiment'].dropna()],
         bins=30, alpha=0.7, label=['Description Sentiment', 'Resolution Sentiment'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Descriptive statistics for sentiment
description_sentiment_stats = ad['Description Sentiment'].dropna().describe()
resolution_sentiment_stats = ad['Resolution Sentiment'].dropna().describe()

# Print the descriptive statistics
print("Description Sentiment Statistics:")
print(description_sentiment_stats)
print("\nResolution Sentiment Statistics:")
print(resolution_sentiment_stats)
```

```python
# Optionally, if you want to see the number of positive, neutral, and negative sentiments
def categorize_sentiment(polarity):
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

ad['Description Sentiment Category'] = ad['Description Sentiment'].apply(categorize_sentiment)
ad['Resolution Sentiment Category'] = ad['Resolution Sentiment'].apply(categorize_sentiment)

# Count occurrences of each category
description_sentiment_counts = ad['Description Sentiment Category'].value_counts()
resolution_sentiment_counts = ad['Resolution Sentiment Category'].value_counts()
```

```python
print("\nDescription Sentiment Category Counts:")
print(description_sentiment_counts)
```

# Customer Segmentation

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select features for segmentation and drop NaN values
features = ad[['Customer Satisfaction Rating', 'Resolution Time (hrs)']].dropna()

# Normalize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
clusters = kmeans.fit_predict(scaled_features)

# Create a new DataFrame for plotting
clustered_ad = features.copy()
clustered_ad['Customer Segment'] = clusters

# Plotting the clusters
plt.figure(figsize=(10, 5))
sns.scatterplot(x=clustered_ad['Customer Satisfaction Rating'], 
                y=clustered_ad['Resolution Time (hrs)'], 
                hue=clustered_ad['Customer Segment'],
                palette='viridis')
plt.title('Customer Segmentation based on Satisfaction and Resolution Time')
plt.xlabel('Customer Satisfaction Rating')
plt.ylabel('Resolution Time (hrs)')
plt.show()
```

```python
# Create a new DataFrame with cluster information
clustered_ad = features.copy()
clustered_ad['Customer Segment'] = clusters

# Print the count of each cluster
cluster_counts = clustered_ad['Customer Segment'].value_counts()
print("Customer Segment Counts:")
print(cluster_counts)

# Descriptive statistics for each cluster
print("\nDescriptive Statistics by Customer Segment:")
for cluster in sorted(clustered_ad['Customer Segment'].unique()):
    cluster_data = clustered_ad[clustered_ad['Customer Segment'] == cluster]
    stats = cluster_data.describe()
    print(f"\nCluster {cluster} Statistics:")
    print(stats)

# Optionally, print the centers of each cluster to understand the characteristics
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
print("\nCluster Centers (Original Scale):")
for i, center in enumerate(cluster_centers):
    print(f"Cluster {i}: Satisfaction Rating = {center[0]:.2f}, Resolution Time (hrs) = {center[1]:.2f}")
```

#### Observation:

Descriptive Statistics by Customer Segment: The statistics for each cluster provide insights into the characteristics of the segments:

Cluster 0 (Medium Satisfaction, Fast Negative Resolution Time):
Average Satisfaction Rating: 3.22 (out of 5), suggesting medium satisfaction levels.
Average Resolution Time: -10.86 hours. The negative value likely indicates an issue with time calculations, perhaps representing tickets resolved before a reference point (like the first response time). This cluster may have faster-than-average resolutions, though the negative times suggest data or logic issues.
Customer Insights: Customers in this segment generally have average satisfaction and quick, albeit possibly improperly logged, resolutions. Generative AI could focus on maintaining or slightly improving satisfaction by addressing possible gaps in understanding negative resolution times.
Cluster 1 (Low Satisfaction, Variable Resolution Time):
Average Satisfaction Rating: 4.07, indicating low satisfaction levels.
Average Resolution Time: 5.61 hours, with high variability (standard deviation of 7.53 hours).
Customer Insights: This segment represents customers who are generally dissatisfied, possibly due to inconsistent or longer resolution times. Enhancing knowledge management processes with Generative AI could help streamline responses and improve satisfaction by providing more consistent and quicker resolutions.
Cluster 2 (High Satisfaction, Moderate Resolution Time):
Average Satisfaction Rating: 1.44, showing high customer satisfaction.
Average Resolution Time: 3.21 hours, indicating moderate resolution times with relatively low variability.
Customer Insights: Customers in this segment are highly satisfied with the service, even if resolution times are not the shortest. This suggests that satisfaction in this group might be driven by the quality of the interaction or resolution rather than just speed. Generative AI can further enhance this group by continuing to personalize and refine the quality of responses.
